```python
colors = {'C':'gray',
         'H':'green',
         'N':'blue',
         'O':'red',
         'B':'orange'}
```

```python
def plot_connex(pref, conf='B'):
    output = f"/scratch/kilian/out/conf_{conf}/{pref}/{pref}_relax.pwo"
    C6H4_C6H4 = read(output, format='espresso-out')
    plt.figure(figsize=(10,7))
    num_to_plot = 1
    for a in C6H4_C6H4:
        if a.symbol != 'H':
            plt.plot(a.position[0],a.position[1],color=colors[a.symbol],
                     marker=f'${str(num_to_plot)}$', markersize=10)
            num_to_plot += 1
    plt.title(pref)
    plt.show()
```

```python
for p in list_of_namesb:
    plot_connex(p)
```

```python
for p in list_of_names_c:
    plot_connex(p,conf='C')
```

```python
def get_vacuum_energy(prefix, conf="B", avg_type='local'):
    
    outfile = f"/scratch/kilian/out/conf_{conf}/{prefix}/avg.dat"
    
    with open(outfile) as f:
        lines = f.readlines()
        
    if avg_type == 'local':
        avg_ind = 1
    else:
        avg_ind = 2

    max_f = float(lines[0].split()[avg_ind])
    min_f = float(lines[0].split()[avg_ind])
    for line in lines:
        V = float(line.split()[avg_ind])
        if V > max_f:
            max_f = V
        if V < min_f:
            min_f = V
            
    print(min_f)
    return round(max_f,4)
```

```python
get_vacuum_energy("C6H4_C6H4")
```
